```python
from sqlalchemy import create_engine

source = create_engine('mysql+mysqlconnector://test:test123@db/test')
```

```python
import pandas as pd
```

```python
import json
with open('unconfirmed-transactions.json') as json_file:
    data = json.load(json_file)
```

```python
df_inter = pd.json_normalize(
            data['txs'],
            meta=[
                    'lock_time', 
                    'ver', 
                    'size', 
                    'weight', 
                    'time', 
                    'hash',
                    'tx_index',
                    'vin_sz',
                    'vout_sz',
                    'relayed_by'
                ]
            )
df_inter
```

```python
df_inter.drop(
    columns=['inputs','out'], 
    axis=1
)
```

```python
df_inputs = pd.json_normalize(
    data['txs'],
    record_path=['inputs'],
    max_level=None,
    meta=['hash'],
)
df_inputs
```

```python
df_outputs = pd.json_normalize(
    data['txs'], 
    record_path=['out'],
     meta=[
                    'lock_time', 
                    'ver', 
                    'size', 
                    'weight', 
                    'time', 
                    'hash',
                    'vin_sz',
                    'vout_sz',
                    'relayed_by'
    ]
)
df_outputs
```

```python
df_join = pd.merge(
    df_inputs, 
    df_outputs,
    how="inner", 
    on="hash", 
    left_on=None, 
    right_on=None, 
    left_index=False, 
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None
)
df_join.head()
```

```python
df_join = pd.merge(
    df_join, 
    df_inter,
    how="inner", 
    on="hash", 
    left_on=None, 
    right_on=None, 
    left_index=False, 
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None
)
df_join.head()
```

```python
print(pd.io.sql.get_schema(df_join.reset_index(), 'test.unconfirmed_transactions', con=source))
```
